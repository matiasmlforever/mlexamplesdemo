Machine learning:  ejemplo de clasificacion

Este clásico ejemplo de "machine learning" busca a través de aprendizaje supervisado poder predecir de qué tipo es la flor Iris observada.

Los atributos son:

Longitud del sépalo en cm (Sepal.Length)
Ancho del sépalo en cm (Sepal.Width)
Longitud del pépalo en cm (Petal.Length)
Ancho del pépalo en cm (Pepal.Width)
Clase (Species):
Iris Setosa
Iris Versicolour
Iris Virginica
Sépalo-Pétalo

![title](img/iris.png)

Carga de librerías y datos.

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_palette('husl')
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

data = pd.read_csv('iris.csv')
```

```python
X = data.drop(['Id', 'Species'], axis=1)
y = data['Species']
# print(X.head())
print(X.shape)
# print(y.head())
print(y.shape)
```

Luego se valida la precisión obtenida para distintos modelos de clasificación K-Neighors y LogisticRegresion.

```python
# Experimentando con N valores distintos
k_range = list(range(1,26))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
    y_pred = knn.predict(X)
    scores.append(metrics.accuracy_score(y, y_pred))
    
plt.plot(k_range, scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of k of k-Nearest-Neighbors')
plt.show()
```

```python
logreg = LogisticRegression()
logreg.fit(X, y)
y_pred = logreg.predict(X)
print(metrics.accuracy_score(y, y_pred))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=5)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Experimentando con N valores distintos
k_range = list(range(1,26))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
    
plt.plot(k_range, scores)
plt.xlabel('Valor de k en KNN')
plt.ylabel('Puntaje Precision')
plt.title('Puntajes de Precision para valores de k en KNN')
plt.show()
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print("Precisión obtenida:")
print(metrics.accuracy_score(y_test, y_pred))
```
